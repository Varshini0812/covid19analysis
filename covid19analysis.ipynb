```python
import pandas as pd
url=r"C:\Users\varsh\Downloads\covid_19_india.csv"
df = pd.read_csv(url)
print(df.info()
print(df.head())
df['Date'] = pd.to_datetime(df['Date'])
print(df.isnull().sum())
```

```python

df['Active_Cases'] = df['Confirmed'] - df['Cured'] - df['Deaths']
countrywise = df.groupby('State/UnionTerritory').sum()[['Confirmed', 'Cured', 'Deaths', 'Active_Cases']].reset_index()
top_10_active_cases = countrywise.nlargest(10, 'Active_Cases')
print(top_10_active_cases)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12,6))
sns.barplot(data=top_10_active_cases, x='State/UnionTerritory', y='Active_Cases', palette='Reds', edgecolor='black')
plt.title('Top 10 Countries by Active COVID-19 Cases')
plt.xticks(rotation=45)
plt.show()
```

```python
# Time series plot for confirmed cases globally
plt.figure(figsize=(12,6))
sns.lineplot(data=df, x='Date', y='Confirmed', hue='State/UnionTerritory', legend=False)
plt.title('COVID-19 Confirmed Cases Over Time (Global)')
plt.xticks(rotation=45)
plt.show()
```

```python
vaccine = vaccine_df[vaccine_df.State!='India']
vaccine
```

```python
vaccine.rename(columns={"Total Individuals Vaccinated":"Total" },inplace = True)
vaccine.head()
```

```python
max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac=max_vac.sort_values('Total',ascending = False)[:5]
max_vac
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(10, 6))
plt.title("Top 5 Vaccinated States in India", size=20)
ax = sns.barplot(data=max_vac, x=max_vac.index, y='Total', linewidth=2, edgecolor="red")
plt.xlabel("States")
plt.ylabel("Total Vaccinations")
plt.xticks(rotation=45) 
plt.show()
```
